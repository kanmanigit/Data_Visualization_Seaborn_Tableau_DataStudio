 # Univariant Analysis
 ## Histogram

```python
import seaborn as sb
sb.distplot(dataset['ssc_p'],kde=True)
```

## Bivariant Analysis: Distribution  (Histogram + Scatter plot)

```python
sb.jointplot(data=dataset,x='ssc_p',y='salary')
plt.show()
```

```python
# Heat map

sb.jointplot(data=dataset,x='ssc_p',y='salary',kind='hex')
plt.show()
```

```python
sb.jointplot(data=dataset,x='ssc_p',y='salary',kind='kde')
plt.show()
```

```python
# Pair plot
```

```python
sb.pairplot(data=dataset,kind="scatter",diag_kind='kde',hue='gender',palette='husl')
plt.show()
```

# Categorical plot
## Stripplot......X axis-categorical data, y axis= Quantitative data

```python
sb.stripplot(data=dataset,x='ssc_b',y='ssc_p',hue='gender')
plt.show()
```

```python
sb.stripplot(data=dataset,x='gender',y='ssc_p',jitter=True)

#Amount of jitter (only along the categorical axis) to apply. This
#    can be useful when you have many points and they overlap, so that
#    it is easier to see the distribution. You can specify the amount
#    of jitter (half the width of the uniform random variable support),
#    or use `True` for a good default.
```

# Swarmplot

```python
sb.swarmplot(data=dataset,x='gender',y='ssc_p',hue='status')
plt.show()
```

# Distribution Observation- Violin plot

```python
sb.violinplot(dataset,x='ssc_p')
plt.show()
```

```python
sb.violinplot(data=dataset,x='gender',y='salary',hue='gender',palette='husl')
```
